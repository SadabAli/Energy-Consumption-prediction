Feature Extraction

```python
# Drop the specified columns
del data["Start time UTC"]
del data["End time UTC"]
del data["Start time IST UTC+05:30"]

# Rename columns with exact names
data.rename(columns={"End time IST UTC+05:30": "DateTime",
                     "Electricity consumption in Odisha ": "Consumption"}, inplace=True)

# Display the first few rows to verify the changes
print(data.head(5))
```

```python
data
```

```python
dataset = data
dataset["Month"] = pd.to_datetime(data["DateTime"]).dt.month
dataset["Year"] = pd.to_datetime(data["DateTime"]).dt.year
dataset["Date"] = pd.to_datetime(data["DateTime"]).dt.date
dataset["Time"] = pd.to_datetime(data["DateTime"]).dt.time
dataset["Week"] = pd.to_datetime(data["DateTime"]).dt.weekday
dataset["Day"] = pd.to_datetime(data["DateTime"]).dt.day_name()
dataset = data.set_index("DateTime")
dataset.index = pd.to_datetime(dataset.index)
```

Data Viz

```python
sns.lineplot(data = dataset , x=dataset['Year'] , y=dataset['Electricity consumption in Odisha'])
plt.title("Electricity consumption in Odisha 2016-2021")
plt.xlabel("Date")
plt.ylabel("Energy in MW")
plt.grid(True)
plt.legend()
```

The graph shows electricity consumption in Odisha from 2016 to 2021. Consumption peaked in 2019, dropped in 2020, and rose again by 2021. This indicates fluctuating energy usage over these years.

```python
plt.figure(figsize=(15,10))
plt.plot(dataset["Electricity consumption in Odisha"])
```

Overall, the graph shows fluctuating energy usage over these years.
This indicates that electricity usage in Odisha has varied significantly, possibly due to factors like economic activities, weather conditions, or policy changes.

```python
from matplotlib import style

fig = plt.figure(figsize = (30,30))

ax1 = fig.add_subplot(611)
ax2 = fig.add_subplot(612)
ax3 = fig.add_subplot(613)
ax4 = fig.add_subplot(614)
ax5 = fig.add_subplot(615)
ax6 = fig.add_subplot(616)

style.use("ggplot")

y_2016 = dataset.loc["2016"]["Electricity consumption in Odisha"].to_list()
x_2016 = dataset.loc["2016"]["Date"].to_list()
ax1.plot(x_2016, y_2016, color= "blue", linewidth= 1.7)

y_2017 = dataset.loc["2017"]["Electricity consumption in Odisha"].to_list()
x_2017 = dataset.loc["2017"]["Date"].to_list()
ax2.plot(x_2017, y_2017, color= "blue", linewidth= 1.7)

y_2018 = dataset.loc["2018"]["Electricity consumption in Odisha"].to_list()
x_2018 = dataset.loc["2018"]["Date"].to_list()
ax3.plot(x_2018, y_2018, color= "blue", linewidth= 1.7)

y_2019 = dataset.loc["2019"]["Electricity consumption in Odisha"].to_list()
x_2019 = dataset.loc["2019"]["Date"].to_list()
ax4.plot(x_2019, y_2019, color= "blue", linewidth= 1.7)

y_2020 = dataset.loc["2020"]["Electricity consumption in Odisha"].to_list()
x_2020 = dataset.loc["2020"]["Date"].to_list()
ax5.plot(x_2020, y_2020, color= "blue", linewidth= 1.7)

y_2021 = dataset.loc["2021"]["Electricity consumption in Odisha"].to_list()
x_2021 = dataset.loc["2021"]["Date"].to_list()
ax6.plot(x_2021, y_2021, color= "blue", linewidth= 1.7)

plt.rcParams["figure.figsize"] = (30, 15)
plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=2.5, wspace=0.4, hspace=0.4)
plt.title("Energy Consumption")
plt.xlabel("Date")
plt.ylabel("Energy in MW")
plt.grid(True, alpha=1)
plt.legend()
```

The pattern in the graph tells us that electricity consumption in Odisha is highly seasonal, with predictable changes across the year. This information is valuable for ensuring that the energy supply meets demand, for identifying opportunities to improve energy efficiency, and for understanding the impact of external events on energy use.
